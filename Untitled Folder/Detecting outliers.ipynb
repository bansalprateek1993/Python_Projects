```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')

df_road = pd.read_csv('datenkostDb.csv')
df_road.head()
```

## Boxplot

```python
%matplotlib inline
ax = sns.boxplot(df_road.Crack_area)

# xtick, label, and title
plt.xticks(fontsize=10)
plt.xlabel('height', fontsize=10)
plt.title('Distribution of crack Area', fontsize=20)
```

```python
%matplotlib inline
ax = sns.boxplot(df_road.Patch_area)

# xtick, label, and title
plt.xticks(fontsize=10)
plt.xlabel('height', fontsize=10)
plt.title('Distribution of crack Area', fontsize=20)
```

## Scatter plot to find relation btw the x and y value

```python
data_riss_fli = pd.DataFrame({'Crack_area': df_road['Crack_area'].tolist() , 'zwriss': df_road['ZWRISS'].tolist()})
ax = sns.scatterplot(x='Crack_area', y='zwriss', data=data_riss_fli, s=50)

# ticks 
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# labels and title
plt.xlabel('Crack Density', fontsize=14)
plt.ylabel('ZEB VALUE', fontsize=14)
plt.title('Relation between density and road goodness value', fontsize=20);
```

```python
data_riss_fli = pd.DataFrame({'Patch_area': df_road['Patch_area'].tolist() , 'zwfli': df_road['ZWFLI'].tolist()})
ax = sns.scatterplot(x='Patch_area', y='zwfli', data=data_riss_fli, s=50)

# ticks 
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# labels and title
plt.xlabel('Patch Density', fontsize=14)
plt.ylabel('ZEB VALUE', fontsize=14)
plt.title('Relation between density and road goodness value', fontsize=20);
```
